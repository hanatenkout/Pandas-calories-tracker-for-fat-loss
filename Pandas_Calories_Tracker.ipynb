```python
eaten_items = []
total_calories = 0
```

```python
def add_item(food_name):
  global total_calories
  food_name = food_name.lower()
  match = df[df['Menu Items'].str.lower().str.contains(food_name)]
  if not match.empty:
    item_row =match.iloc[0]
    item_name = item_row['Menu Items']
    calories = float(item_row['Energy (kCal)'])
    eaten_items.append(item_name)
    total_calories += calories
    print(f"added: {item_name} - {calories} kCal")
  else:
    print("food not found")
```

```python
add_item("paneer")
add_item("fries")
```

```python
def check_tot_calories(limit=750):
  print("\n today's Meals:")
  for item in eaten_items:
    print(f" -{item}")
  print(f"\n Total Calories: {total_calories} kcal")
  if total_calories > limit:
    print("be careful u passed 750 kcal limit!")
  elif total_calories > limit * 0.9 :
    print("u r very close to ur limit")
  else:
    print("u r good to go")
```
